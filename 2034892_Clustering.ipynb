```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from time import time
from sklearn import metrics
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans 
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
```

The Iris dataset was created to quantify the morphologic variation of Iris flowers of three related species: Iris setosa, Iris virginica, and Iris versicolor). There are 50 samples for each species and theres four features measured (length and width of the sepals and petals, in cm). 
In the Iris labels, it contains numbers ranging from 0-2 which indicates the colour of each sample based on their species.

```python
X_data = np.load("Iris_data.npy")
X_labels = np.load("Iris_labels.npy")
colors = np.array(X_labels)

X_sliced = []
Y_sliced = []
Z_sliced = []
A_sliced = []
for i in range(len(X_data)):
  X_sliced.append(X_data[i][0])
  Y_sliced.append(X_data[i][1])
  Z_sliced.append(X_data[i][2])
  A_sliced.append(X_data[i][3])

x = [1,2]

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Scatter Graph of Iris Data")
plt.scatter(X_sliced,Y_sliced, c = colors)
plt.show()

```

KMeans Clustering

```python
kmeans = KMeans(3).fit(X_data)
xmeans = kmeans.predict(X_data)
centers = kmeans.cluster_centers_
plt.scatter(X_data[:,0],X_data[:,1],c=xmeans)
#results = data.predict(,None)
plt.scatter(centers[:,0],centers[:,1],c='red')
#plt.scatter(data)
```

Gaussian Mixture Clustering

```python
kmeans = KMeans(3).fit(X_data)
gMix = GaussianMixture(3,init_params='kmeans').fit(X_data)
labels = gMix.predict(X_data)
means = gMix.means_
plt.scatter(X_data[:,0],X_data[:,1],c=labels)
plt.scatter(means[:,0],means[:,1],c='red')
plt.colorbar()
plt.title("Gaussian Mixture Clustering")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
#print(labels)

```

```python
prob_data = gMix.predict_proba(X_data)
#print(prob_data)
plt.scatter(X_data[:,0],X_data[:,1],c=prob_data[:,0])
plt.title("Component 0")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.colorbar()
```

```python
plt.scatter(X_data[:,0],X_data[:,1],c=prob_data[:,1])
plt.title("Component 1")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.colorbar()
```

```python
plt.scatter(X_data[:,0],X_data[:,1],c=prob_data[:,2])
plt.title("Component 2")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.colorbar()
```
